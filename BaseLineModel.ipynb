```python
import numpy as np 
import pandas as pd 
import seaborn as sns
```

```python
df = pd.read_csv('gurgaon_properties_post_feature_selection.csv')
```

```python
X = df.drop(columns=['price'])
y = df['price']
```

```python
df.sample(2)
```

### so we are planning to apply linear regression here but before that what we will do is all those models which were ordinal encoded we will change them and will perform one hot encoding , then  we will do scaling (standrd scale) , and 3rd thing would be we will apply log transformation to the output col to make it normally distributed

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.svm import SVR
```

```python
columns_to_encode = ['sector', 'balcony', 'agePossession', 'furnishing_type', 'luxury_category', 'floor_category']
```

### we use column transformer to apply transformation and use pipeline from scikit learn jaha pehele transforamtion hori hai and then regrxn model implement hora hai 

```python
# Applying the log1p transformation to the target variable
y_transformed = np.log1p(y)
```

```python
# Creating a column transformer for preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['property_type', 'bedRoom', 'bathroom', 'built_up_area', 'servant room', 'store room']),
        ('cat', OneHotEncoder(drop='first'), columns_to_encode)
    ], 
    remainder='passthrough'
)
```

```python
# Creating a pipeline
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', SVR(kernel='rbf'))
])
```

```python
# K-fold cross-validation
kfold = KFold(n_splits=10, shuffle=True, random_state=42)
scores = cross_val_score(pipeline, X, y_transformed, cv=kfold, scoring='r2')
```

```python
scores.mean()
```

```python
scores.std()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y_transformed,test_size=0.2,random_state=42)
```

```python
pipeline.fit(X_train,y_train)
```

```python
y_pred = pipeline.predict(X_test)
```

```python
y_pred = np.expm1(y_pred)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(np.expm1(y_test),y_pred)
```
